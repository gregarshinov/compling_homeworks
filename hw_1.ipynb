Одним из главных преимуществ regex перед re является поддержка просмотра вперед/назад с нефиксированной длиной (variable length look-arounds). Эта возможность существенно упрощает шаблонизацию последовательностей в специфичных задачах.

### Задача: вернуть мэтч только в случае, если это не телефон банка или агента

```python
import regex
```

```python
texts = ("Контактный телефон банка: 88005555535",
         "Контактный телефон: 88005555535",
         "Контактный телефон агента: 88005555535")
```

```python
expression = regex.compile(r"(?i)контактный телефон(?! банка| агента)")
```

```python
for text in texts:
    print(expression.search(text))
```

Скопируйте текст этой статьи - http://gramota.ru/class/coach/punct/45_177

Приведите его к нижнему регистру, удалите все знаки пунктуации, токенизируйте (любым способом). После этого сделайте стемминг и попробуйте найти по несколько примеров на каждый из видов ошибок (если для какого-то типа не находятся ошибки, так и напишите):
 1. два разных слова ошибочно свелись к одинаковой основе
 2. два словоформы одного слова после стемминга отличаются друг от друга
 3. слово не изменилось после стемминга (слово должно быть русским и длиннее 4 символов)

```python
import requests
```

```python
response = requests.get('http://gramota.ru/class/coach/punct/45_177') # type: requests.Response
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
text_div = soup.select("#taskarea")[0]
```

```python
tokens = [token.lower().strip("_") for token in regex.findall('\w+', text_div.text)]
```

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer("russian")
```

```python
from IPython.core.display import display, HTML
```

```python
stem_map = [{"token": token, "stem": stemmer.stem(token), "id": idx} for idx, token in enumerate(tokens)]
```

```python
header = "<table><thead><thead><tr><th>Токен</th><th>Стем</th></tr><tbody>"
body = "".join(f'<tr><td>{word["token"]}</td><td>{word["stem"]}</td></tr>' for word in stem_map)
footer = "</tbody></table>"
display(HTML(header + body + footer))
```

#### Два разных слова ошибочно свелись к одинаковой основе

```python
for word_1 in stem_map:
    print(word_1)
    for word_2 in stem_map[1:]:
        if word_1["stem"] == word_2["stem"] and word_1["token"] != word_2["token"] and word_1["id"] != word_2["id"]:
            print(f"\t{word_2}")
```

*Ответ*: {'token': 'другого', 'stem': 'друг'} {'token': 'другие', 'stem': 'друг'}

#### Две словоформы одного слова после стемминга отличаются друг от друга

```python
from pymystem3 import Mystem
mystem = Mystem()
```

```python
for word in stem_map:
    word.update(lemma=mystem.lemmatize(word["token"])[0])
```

```python
for word_1 in stem_map:
    print(word_1)
    for word_2 in stem_map[1:]:
        if word_1["lemma"] == word_2["lemma"] and \
        word_1["token"] != word_2["token"] and \
        word_1["stem"] != word_2["stem"]:
            print(f"\t{word_2}")
```

#### Cлово не изменилось после стемминга (слово должно быть русским и длиннее 4 символов)

```python
for word in stem_map:
    if len(word["token"]) > 4 and word["token"] == word["stem"]:
        print(word)
```

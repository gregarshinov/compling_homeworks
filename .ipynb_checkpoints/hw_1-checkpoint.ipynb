Основным модулем для работы с регулярными выражениями в питоне является re. Но есть ещё активно развивающийся regex. Найдите одно любое преимущество regex и опишите его, сопроводив примером (текст, на котором вы демонстрируете отличие, должен быть русским). Пользоваться можно любыми ресурсами, начните поиск с ссылок в семинарской тетрадке. 

Одним из главных преимуществ regex перед re является поддержка просмотра вперед/назад с нефиксированной длиной (variable length look-arounds). Эта возможность существенно упрощает шаблонизацию последовательностей в специфичных задачах.
К примеру:

### Задача: вернуть мэтч только в случае, если это не телефон банка или агента

````python
import regex
````

````python
texts = ("Контактный телефон банка: 88005555535",
         "Контактный телефон: 88005555535",
         "Контактный телефон агента: 88005555535")
````

````python
expression = regex.compile(r"(?i)контактный телефон(?! банка| агента)")
````

````python
for text in texts:
    print(expression.search(text))
````

# Задание 2
Скопируйте текст этой статьи - http://gramota.ru/class/coach/punct/45_177

Приведите его к нижнему регистру, удалите все знаки пунктуации, токенизируйте (любым способом). После этого сделайте стемминг и попробуйте найти по несколько примеров на каждый из видов ошибок (если для какого-то типа не находятся ошибки, так и напишите):
 1. два разных слова ошибочно свелись к одинаковой основе
 2. два словоформы одного слова после стемминга отличаются друг от друга
 3. слово не изменилось после стемминга (слово должно быть русским и длиннее 4 символов)

````python
import requests
````

````python
response = requests.get('http://gramota.ru/class/coach/punct/45_177')
````

````python
from bs4 import BeautifulSoup
````

````python
soup = BeautifulSoup(response.text, 'html.parser')
text_div = soup.select("#taskarea")[0]
````

````python
tokens = [token.lower().strip("_") for token in regex.findall('\w+', text_div.text)]
````

````python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer("russian")
````

````python
from IPython.core.display import display, HTML
````

````python
from pymystem3 import Mystem
mystem = Mystem()
````

````python
stem_map = [{"token": token, "stem": stemmer.stem(token), "id": idx} for idx, token in enumerate(tokens)]
````

````python
for word in stem_map:
    word.update(lemma=mystem.lemmatize(word["token"])[0])
````

````python
header = "<table><thead><thead><tr><th>Токен</th><th>Стем</th></tr><tbody>"
body = "".join(f'<tr><td>{word["token"]}</td><td>{word["stem"]}</td></tr>' for word in stem_map)
footer = "</tbody></table>"
display(HTML(header + body + footer))
````

#### Два разных слова ошибочно свелись к одинаковой основе

````python
for word_1 in stem_map:
    print(word_1)
    for word_2 in stem_map[1:]:
        if word_1["stem"] == word_2["stem"] and \
        word_1["token"] != word_2["token"] and \
        word_1["id"] != word_2["id"] and \
        word_1["lemma"] != word_2["lemma"]:
            print(f"\t{word_2}")
````

**Ответ**: К примеру {'token': 'выделяется', 'stem': 'выделя', 'id': 135, 'lemma': 'выделяться'} {'token': 'выделять', 'stem': 'выделя', 'id': 33, 'lemma': 'выделять'}

#### Две словоформы одного слова после стемминга отличаются друг от друга

````python
for word_1 in stem_map:
    print(word_1)
    for word_2 in stem_map[1:]:
        if word_1["lemma"] == word_2["lemma"] and \
        word_1["token"] != word_2["token"] and \
        word_1["stem"] != word_2["stem"]:
            print(f"\t{word_2}")
````

**Ответ**: к примеру {'token': 'случай', 'stem': 'случа', 'id': 45, 'lemma': 'случай'} {'token': 'случая', 'stem': 'случ', 'id': 410, 'lemma': 'случай'}

#### Cлово не изменилось после стемминга (слово должно быть русским и длиннее 4 символов)

````python
for word in stem_map:
    if len(word["token"]) > 4 and word["token"] == word["stem"]:
        print(word)
````

# Задание 4
Скопируйте текст этой статьи - https://zona.media/news/2019/09/11/zhukov-expertise

Предобработайте текст двумя способами:
1. Разбейте текст на предложения sent_tokenize из nltk, токенизируйте каждое предложение с помощью регулярного выражения (попробуйте разные и выберите лучшее), лемматизируйте токены с помощью pymorphy2

2. предобработайте текст с помощью mystem3 

Ответьте на вопросы:
1. Что лучше разбивает на предложения (mystem или sent_tokenize)?
2. Что лучше токенизирует (mystem или ваша регулярка)? Если mystem, то можно ли улучшить регулярное выражение так, чтобы качество было выше?
3. Что лучше для лемматизации mystem или pymorphy?

````python
from nltk import sent_tokenize
from pymorphy2 import MorphAnalyzer
from pymystem3 import Mystem
from itertools import groupby, zip_longest, chain
morph = MorphAnalyzer()
mystem = Mystem(sentence_markers=True)
from termcolor import colored
````

````python
response = requests.get('https://zona.media/news/2019/09/11/zhukov-expertise')
````

````python
soup = BeautifulSoup(response.text, 'html.parser')
````

````python
text = soup.select("body > div > section > section:nth-child(1) > section > div > div:nth-child(2) > section")[0].text
````

````python
text
````

````python
word_re = regex.compile(r"(\d+|[А-ЯЁа-яё]+(-[А-ЯЁа-яё]+)*)")
````

````python
normalized_sentences_morphy = [
    [
        morph.parse(token[0])[0].normal_form
        for token
        in word_re.findall(sentence)
    ]
    for sentence
    in sent_tokenize(text)
]
for sentence in normalized_sentences_morphy:
    sentence[-1] = sentence[-1] + ' <EOS>'
````

````python

normalized_sentences_mystem = [
    [lemma for lemma in group if word_re.match(lemma)] 
    for is_period, group 
    in groupby(
        mystem.lemmatize(text), 
        lambda token: token == '\\s') 
    if not is_period
]
for sentence in normalized_sentences_mystem:
    sentence[-1] = sentence[-1] + ' <EOS>'
````

````python
len(normalized_sentences_morphy) == len(normalized_sentences_mystem)
````

````python
print("Mystem", "|" ,"nltk.sent_tokenize + Pymorphy\n------------------------------------------")
for my, mo in zip_longest(chain(*normalized_sentences_mystem), chain(*normalized_sentences_morphy)):
    if my != mo:
        print(colored(f"{my} {mo}", 'red'))
    else:
        print(f"{my} {mo}")
````

1. В соответсвии с данным экспериментом, Mystem и sent_tokenize справляются с парсингом предложений одинаково хорошо. Однако в обёртку pymystem3 не включена возможность включать опцию ```-s``` которая выводит маркеры конца предложения. Небольшая модификация pymystem3 позволила это обойти.
2. Mystem токенизирует мягко говоря очень подробно: он включает в список токенов знаки препинания и пробелы. Эта проблема решается добавлением регулярки, которая отфильтровывает лишние (не являющиеся словами и цифрами) токены.
3. Mystem лучше справляется с лемматизацией глаголов (достигать, становиться), а именно различает их категорию вида и указательных местоимений (то). Так же Mystem лучше лемматизирует фамилии, беря в расчет род имени.
